3. What are the different ways to define and call a function in Python?

>  **Defining a Function:**

> A function is defined using the def keyword, followed by the function name, parentheses (), and a colon :. The code block that constitutes the function's logic is indented below the def statement.


```
def function_name(parameter1, parameter2):
    """
    Optional: This is a docstring explaining what the function does.
    """
    # Function body - code to be executed
    result = parameter1 + parameter2
    return result
```

* def keyword: Initiates the function definition.

* function_name: A descriptive name for the function, adhering to Python's naming conventions (lowercase, words separated by underscores).

* (): Parentheses that can optionally contain parameters (input values) the function accepts.

* :: Marks the end of the function signature and the beginning of the indented function body.

* Indentation: All code belonging to the function must be consistently indented.

* return statement (optional): Used to send a value back to the caller. If omitted, the function implicitly returns None.


> **Calling a Function:**

>To execute the code within a defined function, you call it by its name followed by parentheses. If the function expects arguments, they are passed inside the parentheses.



```
# Calling a function without arguments
function_name()

# Calling a function with arguments
returned_value = function_name(argument1, argument2)
```

* function_name(): Calls the function and executes its code.

* argument1, argument2: Values passed to the function's parameters. The order of arguments typically matches the order of parameters in the function definition.

* returned_value: If the function includes a return statement, the returned value can be assigned to a variable.

5. What are iterators in Python and how do they differ from iterables?

> **Iterables:**

> An iterable is an object that can be iterated over, meaning it can return its elements one by one. Common examples of iterables in Python include lists, tuples, strings, and dictionaries. An object is considered iterable if it implements the __iter__ method, which returns an iterator.

> **Iterators:**

> An iterator is an object that represents a stream of data and produces one value at a time using the __next__ method. It maintains an internal state to keep track of the current position during iteration. When there are no more items to return, the __next__ method raises a StopIteration exception.

> **Differences:**

* **Purpose:** An iterable is a container of items that can be iterated over, while an iterator is an object that facilitates the actual iteration process over an iterable.

* **Methods:** Iterables typically implement the __iter__ method. Iterators, on the other hand, implement both the __iter__ method (which returns self) and the __next__ method.

* **State:** Iterators maintain an internal state, remembering where they are in the sequence during iteration. Iterables do not inherently maintain this state.

* **Creation:** An iterator is typically created from an iterable by calling the built-in iter() function or by directly instantiating a class that implements the iterator protocol.

* **Relationship:** All iterators are also iterables (because they implement __iter__), but not all iterables are iterators (e.g., a list is an iterable but not an iterator itself).


Example:



```
# Iterable (a list)
my_list = [1, 2, 3]

# Create an iterator from the iterable
my_iterator = iter(my_list)

# Use the iterator to get elements one by one
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3

# Calling next() again will raise StopIteration
# print(next(my_iterator))
```

6. Explain the concept of generators in Python and how they are defined.

>  A generator is a function that produces a sequence of values lazily, one at a time, rather than storing all values in memory at once. It allows for efficient iteration over large datasets or infinite sequences without loading the entire dataset into memory.


> Generators are defined using the yield keyword within a function, instead of return. When a function contains yield, it automatically becomes a generator function.



```
def count_up_to(n):
    i = 0
    while i < n:
        yield i
        i += 1

# Using the generator
my_generator = count_up_to(5)

for num in my_generator:
    print(num)
```

7. What are the advantages of using generators over regular functions?

> The advantages of using generators over regular functions are primarily related to efficiency and flexibility in handling sequences of data.

* **Memory Efficiency:**

>>  Generators produce values one at a time as they are requested, rather than generating and storing an entire sequence in memory upfront. This is particularly beneficial when dealing with large datasets or potentially infinite sequences, as it significantly reduces memory consumption.

* **Lazy Evaluation:**
>> Values are computed only when they are needed, leading to improved performance, especially when not all elements of a sequence are required. This "on-the-fly" generation avoids unnecessary computations.

* **Handling Infinite Sequences:**
>> Generators can easily represent and iterate over infinite sequences (e.g., Fibonacci numbers, prime numbers) because they do not attempt to store all values in memory.

* **State Preservation and Resumption:**
>> Unlike regular functions that execute entirely and return a single value, generators can pause their execution using the yield keyword and later resume from where they left off, maintaining their internal state between iterations. This allows for stateful iteration and more complex control flow.

* **Composability and Data Pipelines:**
>> Generators can be easily chained together to create powerful and efficient data processing pipelines. This allows for modular and reusable code, enhancing readability and maintainability when performing a series of operations on data.

**Regular Function Example:**



```
def create_large_list(num):
    # Creates and stores the full list in memory
    return [i for i in range(num)]

# This will consume a lot of memory for a large `num`
my_list = create_large_list(1000000)
```

**Generator Example:**



```
def generate_numbers(num):
    # Yields one number at a time, preserving memory
    for i in range(num):
        yield i

# The numbers are generated as needed, not all at once
my_generator = generate_numbers(1000000)
```

9. Explain the purpose and usage of the 'map() function in Python.

> The map() function in Python serves the purpose of applying a specified function to each item within an iterable (such as a list, tuple, or set) and returning an iterator that yields the results. It provides a concise and efficient way to transform all elements of an iterable without explicitly writing a for loop.

> The map() function takes two main arguments:

* function:

 The function to be applied to each item of the iterable. This can be a built-in function, a user-defined function, or a lambda function for inline operations.

* iterable:

The sequence of items to which the function will be applied.

Example:



```
# Define a list of numbers
numbers = [1, 2, 3, 4, 5]

# Define a function to square a number
def square(x):
    return x * x

# Use map() to apply the square function to each number
squared_numbers_map_object = map(square, numbers)

# Convert the map object to a list to view the results
squared_numbers_list = list(squared_numbers_map_object)

print(squared_numbers_list)
# Output: [1, 4, 9, 16, 25]

# Using a lambda function with map()
doubled_numbers = list(map(lambda x: x * 2, numbers))
print(doubled_numbers)
# Output: [2, 4, 6, 8, 10]
```

0. What is the difference between `map(), `reduce(), and `filter() functions in Python?

> The map(), reduce(), and filter() functions in Python are higher-order functions used for processing iterables, but they differ in their purpose and the type of output they produce:

> i> **map(function, iterable):**

* Purpose: Applies a given function to each item in an iterable and returns an iterator that yields the results.

* Output: An iterator containing the transformed elements, maintaining the same number of elements as the original iterable. Example: Doubling each number in a list.


```
  # map function

    numbers = [1, 2, 3, 4]
    doubled_numbers = map(lambda x: x * 2, numbers)
    # doubled_numbers will yield 2, 4, 6, 8
```

> ii> **reduce(function, iterable[, initial]):**

* Purpose: Applies a rolling computation to sequential pairs of elements in an iterable, reducing the iterable to a single cumulative value. It is found in the functools module in Python 3.

* Output: A single value representing the aggregated result of the computation.
Example: Summing all numbers in a list.



```
  # reduce function

   from functools import reduce
    numbers = [1, 2, 3, 4]
    sum_of_numbers = reduce(lambda x, y: x + y, numbers)
    # sum_of_numbers will be 10
```


> iii> **filter(function, iterable):**

* Purpose: Constructs an iterator from elements of an iterable for which the function returns True. The function must return a boolean value.

* Output: An iterator containing only the elements from the original iterable that satisfy the condition, potentially having fewer elements than the original.
Example: Filtering out odd numbers from a list.




```
   # filter function

   numbers = [1, 2, 3, 4, 5]
    even_numbers = filter(lambda x: x % 2 == 0, numbers)
    # even_numbers will yield 2, 4
```

````python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two_generator(max_exponent):

    if not isinstance(max_exponent, int) or max_exponent < 0:
        raise ValueError("max_exponent must be a non-negative integer.")

    for i in range(max_exponent + 1):
        yield 2 ** i

powers_of_two_generator(10)
````

````python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
    """
    Reads a file line by line and yields each line as a string.

    Args:
        file_path (str): The path to the file to be read.

    Yields:
        str: Each line from the file.
    """
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip('\n')  # Yield the line, removing the newline character
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
# Create a dummy file for demonstration
with open("example.txt", "w") as f:
    f.write("This is line 1.\n")
    f.write("This is line 2.\n")
    f.write("This is line 3.\n")

# Iterate over the lines using the generator
for line in read_file_line_by_line("example.txt"):
    print(line)

# Example with a non-existent file
for line in read_file_line_by_line("non_existent_file.txt"):
    print(line)
````

````python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

celsius_temperatures = [0, 10, 20, 30, 40, 50]

celsius_to_fahrenheit = lambda c: (c * 9/5) + 32

fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print("Celsius temperatures:", celsius_temperatures)
print("Fahrenheit temperatures:", fahrenheit_temperatures)
````
